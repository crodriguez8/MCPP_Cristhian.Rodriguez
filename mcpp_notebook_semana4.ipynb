```python
# Construir lista de 10000 números aleatorios (entre 0 y 9)

import random  # importar módulo (librería)

N = 10000  
random_numbers = []  # Inicializar lista vacía
for i in range(N):  # Para i en [0, 9999]. (10000 veces) 
    random_numbers.append(random.randint(0,9))  # Añadir a la lista "random_numbers" un entero entre 0 y 9
```

```python
random_numbers
```

```python
print(random_numbers)
```

```python
len(random_numbers)
```

```python
# Construir lista de 100 números aleatorios (entre 0 y 9)

import random  # importar módulo (librería)

N = 100
random_numbers2 = []  # Inicializar lista vacía
for i in range(N):  # Para i en [0, 99]. (100 veces) 
    random_numbers2.append(random.randint(0,9))  # Añadir a la lista "random_numbers" un entero entre 0 y 9
    print(random_numbers2)
    
```

¿Cómo hallar la frecuencia de cada número del 0 al 9?

```python
count = []  # Inicializar lista
for x in range(0,10):
    frecuencia = random_numbers.count(x)  # contar ocurrencias de x en lista "random_numbers"
    count.append(frecuencia)  # agregar "frecuencia" a lista "count"
```

```python
count
```
